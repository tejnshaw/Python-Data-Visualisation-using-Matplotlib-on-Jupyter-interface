```python
haiti=df.loc['Haiti', year]
print(haiti)
```

```python
haiti.plot() # to check absurd increase in immigration in Haiti
```

```python
haiti.plot(kind='line')
plt.title('Immigration from Haiti')
plt.xlabel('Years')
plt.ylabel('No. of Immigrants')
plt.text(20, 6000, '2010 Earthquake')
plt.show()
```

```python
#sort values by top 5 immigrats and plotting them
df.sort_values(by='Total', ascending=False, axis=0, inplace=True)
df2=df.head(5)
df2=df2.transpose()
df3=df2
df3.drop(['Continent','Region','DevName','Total'], axis=0,inplace=True)
df3
```

```python
df3.plot(kind='line', figsize=(14, 8)) # pass a tuple (x, y) size

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```
